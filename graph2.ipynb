```python
import networkx as nx
g=nx.Graph()
g.add_nodes_from([0,1,2,3,4])
g.add_node("a")
g.add_node("b")
g.add_node("c")
g.add_node("d")

print(g)
nx.draw(g, with_labels = True)
```

```python
g.add_edges_from([
    (0,"a"),
    (0,"b"),
    (1,"b"),
    (1,"c"),
    (2,"c"),
    (2,"d"),
    (3,"d"),
    (4,"d")
])
print(g)
nx.draw(g, with_labels = True)
```

```python

a=nx.bfs_tree(g,1)
for i in g.nodes:
    if i not in a:
        b="graph is not connected"
        break
    else:
        b="graph is connected" 
        
print(b)
```

```python
c=nx.is_connected(g)
if c:
    print("graph is connected")
```

```python
d=nx.is_bipartite(g)
print(d)
```

```python
x,y = nx.bipartite.sets(g)
print(x,y)
```
